# 1. What is a Jupyter Notebook?

A **Jupyter Notebook** is an interactive document that combines:
- **text**
- **code**
- **output**
- **visualizations**

This makes it very useful for:
- teaching
- data analysis
- machine learning
- experiments
- step-by-step explanations

A notebook is usually saved as a file with the extension:

```text
.ipynb
```

That extension stands for **IPython Notebook**, although today Jupyter supports more than just IPython.

## Why do people like notebooks?

Notebooks are popular because they let you:
- write an explanation in plain language
- run code directly below the explanation
- immediately see results
- mix theory and practice in one place

This is especially useful in education because students can:
- read the concept
- run an example
- modify the example
- observe what changes

# 2. What is Google Colab?

**Google Colab** (or **Colaboratory**) is a **hosted Jupyter Notebook service**.

In simple words:
- a Jupyter notebook normally can run on your own computer
- Colab lets you run a Jupyter-style notebook **in your browser**
- you do not need to install Python locally just to get started

Colab is very popular for beginners because it is easy to access and quick to use.

## Jupyter Notebook vs Google Colab

| Feature | Jupyter Notebook | Google Colab |
|---|---|---|
| Where it runs | Usually on your computer or server | In the browser |
| Setup required | Often yes | Usually no setup to start |
| File format | `.ipynb` | `.ipynb` |
| Good for | Local work, custom environments | Fast start, cloud compute, sharing |
| Collaboration | Possible, but depends on setup | Easy sharing like Google Docs |

**Important:** Colab uses the **same notebook file format** (`.ipynb`), which makes it easy to move files between tools.

# 10. Useful notebook actions

Typical actions in notebook tools include:
- add a cell above or below
- change cell type
- run a cell
- restart runtime / kernel
- clear outputs
- save notebook

Typical keyboard ideas:
- **Shift + Enter**: run current cell and move to next

# 11. Using Colab via the website

# 15. How to use notebooks in VS Code

VS Code can open and edit `.ipynb` files.

## Typical setup
Students usually install:
- **VS Code**
- **Python extension**
- **Jupyter extension**

Then they can:
- open a notebook file
- edit Markdown and code cells
- run cells locally
- inspect outputs
- debug code more easily than in some browser-only workflows

## Why use VS Code for notebooks?

VS Code is useful when students want:
- local files and folders
- stronger code editing support
- terminal access
- debugging
- Git integration
- a smoother transition from notebooks to `.py` scripts

# 16. Colab and VS Code: practical workflow

For most teaching situations, the easiest workflow is:

## Option A: Start in Colab
1. Create notebook in Colab.
2. Download the notebook as `.ipynb`.
3. Open the same file in VS Code.

## Option B: Start in VS Code
1. Create or edit notebook in VS Code.
2. Save the `.ipynb` file.
3. Upload it to Colab or open it from Drive/GitHub in Colab.

Because both tools use the `.ipynb` format, moving files between them is straightforward.

# 17. Can Colab use a local runtime?

Yes, Colab can connect to a **local runtime**.

This means:
- the **Colab browser interface** stays in the browser
- the **code execution** can happen on your own machine

This is more advanced than the beginner browser workflow, but it is useful to know.

## Important caution
Connecting Colab to a local runtime means notebook code may access local machine resources.
Only do this with notebooks you trust.

# 20. Example: simple plot
A notebook can also display visual output.

# 21. Common beginner mistakes

## Mistake 1: Running cells in the wrong order
Result: variables may not exist, or values may be outdated.

## Mistake 2: Forgetting to run a cell
Result: no output appears.

## Mistake 3: Editing a cell but not rerunning it
Result: the output still reflects old code.

## Mistake 4: Losing work in a temporary runtime
Result: installed packages or uploaded temporary files may disappear after restart.

## Mistake 5: Confusing Markdown and code cells
Result: text may produce syntax errors if placed in a code cell.
